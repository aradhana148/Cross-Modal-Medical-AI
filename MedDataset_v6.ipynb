````python


def append_to_json(file_path, data):
  """Appends data to a JSON file.

  Args:
    file_path: The path to the JSON file.
    data: The data to append (can be a dictionary or a list).
  """
  with open(file_path, 'r+') as f:
    # Load the existing data
    try:
      file_data = json.load(f)
    except json.JSONDecodeError:
      #file_data = []  # If file is empty, initialize with an empty list
      print("File reading exception is happening WARNING")

    # Append the new data
    if isinstance(file_data, list):
      file_data.append(data)  # Append to list
    elif isinstance(file_data, dict):
      file_data.update(data)  # Update dictionary
    else:
      raise ValueError("JSON file should contain a list or a dictionary.")

    # Reset file pointer to the beginning
    f.seek(0)

    # Write the updated data back to the file
    json.dump(file_data, f, indent=4)

    # Truncate the file to remove any extra data
    f.truncate()


````

````python
def saveDataLog(filename, datatosave):
  with open(filename, 'w') as f:
    json.dump(datatosave, f, indent=4)
````

````python
def generate(myprompttext1, video, myprompttext2):
    vertexai.init(project="probable-splice-440907-h2", location="us-central1")
    model = GenerativeModel(        "gemini-1.5-flash-002",    )
    responses = model.generate_content(
        [myprompttext1, video, myprompttext2],
        generation_config=generation_config,
        safety_settings=safety_settings,
        stream=True,    )
    return responses
````
